# Importing Basic Liabraris & Dataset for analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("D:\\LetsGrowMore\\Task3\\globalterrorismdb_0718dist.csv", encoding= 'latin', low_memory=False)
```

# Data Insights & Data Cleaning

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
final_df = df[['iyear','imonth', 'iday','extended','country_txt','region_txt','city','success','suicide','attacktype1_txt',
             'targtype1_txt','gname','weaptype1_txt','nkill']]
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df.columns
```

```python
final_df.rename(columns={'iyear' : 'year', 'imonth' : 'month', 'iday' : 'day', 'country_txt' : 'country', 
                         'region_txt' : 'region',  'attacktype1_txt' : 'attack type', 'targtype1_txt': 'target type',
                         'gname' : 'gang name', 'weaptype1_txt' : 'weapon type', 'nkill' : 'no of kills'}, inplace = True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df.describe()
```

```python
final_df.describe(include='O')
```

```python
final_df.isnull().sum()
```

```python
final_df['no of kills'] = final_df['no of kills'].fillna(0)
```

```python
final_df['city'] = final_df['city'].fillna('unknown')
```

```python
final_df.isnull().sum()
```

```python
final_df.duplicated().sum()
```

```python
final_df.drop_duplicates(inplace= True)
```

```python
final_df.shape
```

```python
final_df['no of kills'] = final_df['no of kills'].astype(int)
```

```python
# Exporting Cleaned Dataset for making analysis with different visualization tools like PowerBI  
final_df.to_csv("D:\\LetsGrowMore\\Task3\\TerroristDataset.csv")
```
